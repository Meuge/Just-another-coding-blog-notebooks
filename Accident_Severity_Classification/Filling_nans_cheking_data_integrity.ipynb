#### Information gathered from _lat, long, postal code, county, state, and city_ 

When we made an analysis of the location data provided, it could be observed that 2126 cases were not correct, that is, one or more of them was not congruent with the associations of the others, that is, county, city, state, or zip code. That is why it was decided to carry out a check with the USPS API, United Post Service, to verify how reliable and complete the data was, based on the latitude and longitude entered. **Why USPS?**  Because it was who registered the postal code in the USA in 1963, and who carried out the postal code expansions.  After forming our query structure with USPS, we will proceed to query the BING API, and we will take into account the points near a neighborhood (square). If any of them at a small distance correspond with the given zip code, we take it as the case of a possible match.To analyze the possible matches of each entry, we give priority to what is entered in the dataset, verifying with our list of probable candidates. First, we start with zip code, and then we go on to analyze state, county, and city. We used the same procedure for the 880 Zipcode and 83 City NANs.

```python
import requests
from time import sleep
#WE BELIEVE THAT USPS IT'S THE MOST RELIABLE SOURCE TO VERITY THE INTEGRITY OF THE DATA
#USPS BRINGS US CITY, STATE, ZIPCODE,BUT NOT COUNTY
def usps_zipcode_api(zipcode:str):
  headers = {
  'authority': 'm.usps.com',
  'pragma': 'no-cache',
  'cache-control': 'no-cache',
  'accept': 'application/json, text/javascript, */*; q=0.01',
  'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.138 Safari/537.36',
  'x-requested-with': 'XMLHttpRequest',
  'sec-fetch-site': 'same-origin',
  'sec-fetch-mode': 'cors',
  'sec-fetch-dest': 'empty',
  'referer': 'https://m.usps.com/m/ZipLookupAction?search=zip',
  'accept-language': 'en,es-419;q=0.9,es;q=0.8',
  'cookie': 'JSESSIONID=0000SL-cA90BnSOvHmSAQ_cW-ZW:1bpccqlcc; NSC_psjhjo-n_443=ffffffff3b2236d945525d5f4f58455e445a4a4212d3',
  }
  params = (('mode', '2'),('tZip', zipcode),('jsonInd', 'Y'),)
  response = requests.get('https://m.usps.com/m/QuickZipAction', headers=headers, params=params)
  location = {}
  for address in response.json()['addresses']:
    location.setdefault(address['state'], set()),
    location[address['state']].add(address['city'])
  for k in location.keys():
    location[k] = sorted(location[k])
  return location
```

```python
# BING API TO SOLVE CONFLICTED ISSUES
import requests, json

def bing_data(lat, lng, use_locality=False):
	headers = {
    'authority': 'dev.virtualearth.net',
    'pragma': 'no-cache',
    'cache-control': 'no-cache',
    'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.138 Safari/537.36',
    'dnt': '1',
    'accept': '*/*',
    'sec-fetch-site': 'cross-site',
    'sec-fetch-mode': 'no-cors',
    'sec-fetch-dest': 'script',
    'referer': 'https://dev.virtualearth.net/',
    'accept-language': 'en-US,en;q=0.9',
	}

	params = (
    ('o', 'json'),
    ('jsonp', 'Microsoft.Maps.NetworkCallbacks.f574e6'),
    ('key', 'AoltDrgPXkLphU6xf7mwndXly_XEVrnForv0FrqgPHHXwIf-oaB3bfErNvlSYVXc'),
    ('includeEntityTypes', ''),
    ('inclnb', '0'),
    ('incl', ''),
    ('ur', 'ar'),
    ('c', 'en-US'),
	)

	response = requests.get(f'https://dev.virtualearth.net/REST/v1/Locations/{lat},{lng}', headers=headers, params=params)

	data = json.loads(response.text[response.text.find('{'):response.text.rfind('}') + 1])
	state = data['resourceSets'][0]['resources'][0]['address']['adminDistrict']
	county = data['resourceSets'][0]['resources'][0]['address']['adminDistrict2'][:-4].upper()
	zipcode = data['resourceSets'][0]['resources'][0]['address']['postalCode']
	if use_locality:
		locality = data['resourceSets'][0]['resources'][0]['address']['locality'] # WE GENERALLY DONT USE IT BECAUSE USPS HAS A CLEARER DEFINITION OF CITY
		return (state, county, locality, zipcode)
	return (state, county, zipcode)
```

```python
import requests, json

def bing_timezone_api(lat, lng):
  headers = {
    'authority': 'dev.virtualearth.net',
    'pragma': 'no-cache',
    'cache-control': 'no-cache',
    'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.138 Safari/537.36',
    'dnt': '1',
    'accept': '*/*',
    'sec-fetch-site': 'cross-site',
    'sec-fetch-mode': 'no-cors',
    'sec-fetch-dest': 'script',
    'referer': 'https://dev.virtualearth.net/',
    'accept-language': 'en-US,en;q=0.9',
  }

  params = (
      ('o', 'json'),
      ('jsonp', 'Microsoft.Maps.NetworkCallbacks.f77b6a'),
      ('key', 'AoltDrgPXkLphU6xf7mwndXly_XEVrnForv0FrqgPHHXwIf-oaB3bfErNvlSYVXc'),
      ('includeEntityTypes', 'Neighborhood,PopulatedPlace,AdminDivision1,AdminDivision2,CountryRegion'),
      ('inclnb', '1'),
      ('incl', 'entityId'),
      ('c', 'en-US')
  )
  response = requests.get(f'https://dev.virtualearth.net/REST/v1/TimeZone/{lat},{lng}', headers=headers, params=params)
  data =  json.loads(response.text[response.text.find('{'):response.text.rfind('}') + 1])
  timezones = [{k:v} for resourceSets in data['resourceSets'] for resources in resourceSets['resources'] for k, v in resources['timeZone'].items() if k in {'genericName', 'abbreviation', 'windowsTimeZoneId'}]
  
  timezone = set()
  timezone_format = {'Eastern Standard Time': 'EST', 'Central Standard Time' : 'CST' , 'Mountain Standard Time' : 'MST', 'Pacific Standard Time' : 'PST'}
  for t in timezones:
    if 'abbreviation' in t.keys():
      timezone.add(t['abbreviation'])
    elif 'genericName' in t.keys() and t['genericName'].find('Standard Time') > 0:
      for k,v in timezone_format.items():
        if t['genericName'].find(k) >= 0:
          timezone.add(v)
          break
    elif 'windowsTimeZoneId' in t.keys() and t['windowsTimeZoneId'].find('Standard Time') > 0:
      for k,v in timezone_format.items():
        if t['windowsTimeZoneId'].find(k) >= 0:
          timezone.add(v)
          break
    else:
      print(f'ERROR TIMEZONE {timezone}')
      raise f'ERROR TIMEZONE {timezone}'

  return sorted(timezone)
```

```python
# FILL BY WEATHER CONDITION & SHORTER
threshold = 5
temperature_update = []
sample_filler = {}
for temperature, weather, zipcode, city, county, state in zip(us_accident_data['Temperature(F)'], us_accident_data.Weather_Condition, us_accident_data.Zipcode, us_accident_data.City, us_accident_data.County, us_accident_data.State):
  if not np.isnan(temperature) and type(weather) == str:
    # CONSIDERING WEATHER CONDITION
    sample_filler.setdefault(weather+state, [0,0])
    sample_filler.setdefault(weather+state+county, [0,0])
    sample_filler.setdefault(weather+state+county+city, [0,0])
    sample_filler.setdefault(weather+zipcode, [0,0])
    # ADD WINDSPEED
    # CONSIDERING WEATHER CONDITION
    sample_filler[weather+state][0] += temperature
    sample_filler[weather+state+county][0] += temperature
    sample_filler[weather+state+county+city][0] += temperature
    sample_filler[weather+zipcode][0] += temperature
    # COUNT CASES
    # CONSIDERING WEATHER CONDITION
    sample_filler[weather+state][1] += 1
    sample_filler[weather+state+county][1] += 1
    sample_filler[weather+state+county+city][1] += 1
    sample_filler[weather+zipcode][1] += 1
  if not np.isnan(temperature):  
    # CONSIDERING ONLY LOCATION
    sample_filler.setdefault(state, [0,0])
    sample_filler.setdefault(state+county, [0,0])
    sample_filler.setdefault(state+county+city, [0,0])
    sample_filler.setdefault(zipcode, [0,0])
    # ADD WINDSPEED
    # CONSIDERING ONLY LOCATION
    sample_filler[state][0] += temperature
    sample_filler[state+county][0] += temperature
    sample_filler[state+county+city][0] += temperature
    sample_filler[zipcode][0] += temperature
    # COUNT CASES
    # CONSIDERING ONLY LOCATION
    sample_filler[state][1] += 1
    sample_filler[state+county][1] += 1
    sample_filler[state+county+city][1] += 1
    sample_filler[zipcode][1] += 1

# FIND MEAN WHEN THE CASES ARE GREATER THAN THE THRESHOLD
mean_filler = {key:value[0]/value[-1] for key, value in sample_filler.items() if value[-1] >= threshold}


print()
print(f'STARTING CONDITION WITH THRESHOLD OF {threshold}')
print('nan cases', len(us_accident_data['Temperature(F)'][us_accident_data['Temperature(F)'].isna()]), 'nan percentage', 100 * len(us_accident_data['Temperature(F)'][us_accident_data['Temperature(F)'].isna()]) / len(us_accident_data['Temperature(F)']))
print()
for temperature, weather, zipcode, city, county, state in zip(us_accident_data['Temperature(F)'], us_accident_data.Weather_Condition, us_accident_data.Zipcode, us_accident_data.City, us_accident_data.County, us_accident_data.State):
  if np.isnan(temperature):
    if type(weather) == str: # TRY FILLING WITH WEATHER CONDITION
      temperature = mean_filler.get(weather+zipcode, np.nan) if np.isnan(temperature) else temperature
      temperature = mean_filler.get(weather+state+county+city, np.nan) if np.isnan(temperature) else temperature
      temperature = mean_filler.get(weather+state+county, np.nan) if np.isnan(temperature) else temperature
      temperature = mean_filler.get(weather+state, np.nan) if np.isnan(temperature) else temperature
    temperature = mean_filler.get(zipcode, np.nan) if np.isnan(temperature) else temperature
    temperature = mean_filler.get(state+county+city, np.nan) if np.isnan(temperature) else temperature
    temperature = mean_filler.get(state+county, np.nan) if np.isnan(temperature) else temperature
    temperature = mean_filler.get(state, np.nan) if np.isnan(temperature) else temperature
  temperature_update.append(temperature)
temperature_update_nan = len([xi for xi in temperature_update if np.isnan(xi)])
print('nan cases', temperature_update_nan, 'nan percentage', 100 * temperature_update_nan / len(temperature_update))
print()
```
